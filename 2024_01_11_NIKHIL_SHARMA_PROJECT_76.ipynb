#### Activity 1: Plot the Equation

In this activity, you need to plot the equation on the graph for the roller coaster.

$$y =  -0.000833 (x^3 + 12x^2 - 580x -1200) $$


Where,

- $y$ is the height of the roller coaster from the ground.

- $x$ is the horizontal length of the roller coaster


**Input**
* The input array should range from -30 to 30

**Steps**
1. Create the function to form the curve of the Roller Coaster with the equation given which takes an array as and returns the curve for the array.
2. Create an input array of 1000 numbers ranging from -30 to 30.
3. Call the function created for the curve with the input array created and save the output array returned in a new variable.
4. Plot the graph to display the path of the Roller Coaster using the input and output array.

```
# Plot the curvature of the coaster using between x = -30 and x = 30 only.

# Import the modules.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
# Create the function to return the curve of the given equation for the input array/list

def func(x):
  y = -(0.000833)*(x**3 + 12*x**2 - 580*x - 1200)
  return y
# Create an input array
x = np.linspace(-30, 30, 10**5)
# Call the function created for curve on the input array and save it in a new variable
f = func(x)
# Plot the graph for the Roller Coaster
plt.figure(figsize = (10,4))
plt.plot(x,f)
# Plot a line x-axis, y-axis at (0,0)
plt.axhline(0, color = 'black')
plt.axvline(0, color ='black')
# Create gridlines
plt.grid()
# Label both the axes
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#### Activity 2: Calculate and Plot the Derivative of the Equation

In this activity, you need to calculate the rate of change from the ground to
the top of the coaster using the derivative (or differentiation) equation and plot the graph for the same.

\begin{equation}
\frac{\delta y}{\delta x} = \frac{f(x + h) - f(x)}{h}
\end{equation}

$$ $$
\begin{equation}
\therefore \frac{\delta y}{\delta x} = \frac{y_i - y_{i-1}}{x_i - x_{i-1}}
\end{equation}


**Input**
* Use the input created in Activity 1


**Steps**
1. Create a function to calculate the derivative which takes two arrays - input array for x-axis and curvature array for y axis as and returns the derivative using the equation given for  $\delta y$.
2. Call the function created to calculate derivative with the input and output array created in the Activity 1 and save the derivative array returned in a new variable.
4. Plot the graph to display the rate of change in the path of the Roller Coaster using the input and derivative array.


```
# Find the derivative and create the graph for the derivative of the input array created in Activity 1
def derivative(x,y):
  l = []
  for i in range(len(y)):#0 to len(y)-2
    der = (y[i]-y[i-1])/(x[i]-x[i-1])
    l.append(der)
  return l
# Create the function for finding the derivative of the input array
#def derivative_func(x):
#  y = -(0.000833)*(3*x**2 + 24*x - 580)
# return y
y2 = derivative(x,f)
# Plot the derivative graph
plt.figure(figsize = (10,4))
plt.plot(x,y2)
# Plot a line x-axis,y-axis at (0,0)
plt.axhline(0, color = 'black')
plt.axvline(0, color = 'black')
# Create gridlines
plt.grid()
# Labelling the axes
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#### Activity 3: More Derivatives - Derivative of Log

In this acivity you have to plot the derivative of different logarithms
$$y=[\log_{10}x,\log_{10}2x,\log_{10}3x,\log_{10}5x,\log_{10}10x]$$

**Steps**
1. Create an input array of integers 1-100.
2. Calculate the different logarithms given in the question with respect to the input array and save them each in  another array/list.
3. Plot the graph for log terms.
4. Calculate the derivatives for the log terms with respect to the input array.
5. Plot the graph of the derivative of the log terms.

```
# Import required libraries.


# Create input for x-axis.
x = np.arange(1,101)

# Evaluate log(x) with base 10 for corresponding values of x.
y1 = np.log10(x)
# Evaluate log(2x) with base 10 for corresponding values of x.
y2 = np.log10(2*x)
# Evaluate log(3x) with base 10 for corresponding values of x.
y3 = np.log10(3*x)
# Evaluate log(5x) with base 10 for corresponding values of x.
y4 = np.log10(5*x)
# Evaluate log(10x) with base 10 for corresponding values of x.
y5 = np.log10(10*x)
```

Create the graph for the different logarithm of the input array.

```
# Plot the graphs for the logarithm
plt.figure(figsize = (10,4))
# Add title to the graph
plt.title("Plotting logs")
# plot the graphs
plt.plot(x, y1, color = 'red')
plt.plot(x, y2, color = 'orange')
plt.plot(x, y3, color = 'blue')
plt.plot(x, y4, color = 'black')
plt.plot(x, y5, color = 'green')
# Labelling the axes
plt.xlabel("x-axis")
# Show a legend on the plot
plt.ylabel("y-axis")
# Show the plot
plt.show()
```

Create the graph for the different derivatives of the input array.

```
# Plot the graphs for the derivatives of the logarithms
y1_prime = derivative(x,y1)
y2_prime = derivative(x,y2)
y3_prime = derivative(x,y3)
y4_prime = derivative(x,y4)
y5_prime = derivative(x,y5)
# Add title to the graph
plt.title("Plotting derivative of logs")
# Plot the graphs
plt.plot(x, y1_prime, label = "log10(x)",color = 'red')
plt.plot(x, y2_prime, label = "log10(2*x)",color = 'orange')
plt.plot(x, y3_prime, label = "log10(3*x)",color = 'yellow')
plt.plot(x, y4_prime, label = "log10(5*x)",color = 'blue')
plt.plot(x, y5_prime, label = "log10(10*x)",color = 'green')
# Labelling the axes
plt.xlabel("x-axis")
plt.ylabel("y-axis")
# Show a legend on the plot
plt.legend()
# Show the plot
plt.show()
```
