#### Activity 1:  Plot a Graph with $y = \{\log_{10} x, \log_{10} 2x, \log_{10} 3x, \log_{10} 5x, \log_{10} 10x \}$


1. Take list `x` and store number 1 to 99.

2. Calculate corresponding values for $y = \{\log_{10} x, \log_{10} 2x, \log_{10} 3x, \log_{10} 5x, \log_{10} 10x \}$
and store them in variables `y1`, `y2`,`y3`, `y4`, and `y5` respectively.

3. Plot the graph and make your observations.

```
import numpy as np
import math
import matplotlib.pyplot as plt

x = []
for i in range(1,100):
  x.append(i)
y1_new = []
for i in range(len(x)):
  y1_new.append(math.log(x[i],10))
y2_new = []
for i in range(len(x)):
  y2_new.append(math.log(2*x[i],10))
y3_new = []
for i in range(len(x)):
  y3_new.append(math.log(3*x[i],10))
y4_new = []
for i in range(len(x)):
  y4_new.append(math.log(5*x[i],10))
y5_new = []
for i in range(len(x)):
  y5_new.append(math.log(10*x[i],10))

plt.title("Plots of Logs")
plt.plot(x,y1_new,color = 'red')
plt.plot(x,y2_new,color = 'orange')
plt.plot(x,y3_new,color = 'yellow')
plt.plot(x,y4_new,color = 'blue')
plt.plot(x,y5_new,color = 'green')
plt.xlabel('Values')
plt.ylabel('log of Values')
plt.show()
```

**Q:** What are your observations?

**A:** when the expression under log base 10 of (someething x or 2x) becomes 1 then the graph cuts at zero, all the other plots are just a constant+log10(x) plot

#### Activity 2: Plot a graph with $y=\{\log_{1.1} x,\log_2 x,\log_3 x,\log_5 x,\log_{10} x\}$

Calculate corresponding values for $y=\{\log_{1.1} x,\log_2 x,\log_3 x,\log_5 x,\log_{10} x\}$
and store them in variables `y1`, `y2`,`y3`, `y4`, and `y5` respectively.



```
# Evaluate log(x) with base 1.1 for corresponding values of x.
y1_new = []
for i in range(len(x)):
  y1_new.append(math.log(x[i],1.1))
y2_new = []
for i in range(len(x)):
  y2_new.append(math.log(x[i],2))
y3_new = []
for i in range(len(x)):
  y3_new.append(math.log(x[i],3))
y4_new = []
for i in range(len(x)):
  y4_new.append(math.log(x[i],5))
y5_new = []
for i in range(len(x)):
  y5_new.append(math.log(x[i],10))
plt.title("Plots of Logs")
plt.plot(x,y1_new,color = 'red')
plt.plot(x,y2_new,color = 'orange')
plt.plot(x,y3_new,color = 'yellow')
plt.plot(x,y4_new,color = 'blue')
plt.plot(x,y5_new,color = 'green')
plt.xlabel('Values')
plt.ylabel('log of Values')
plt.show()
```

**Q:** What are your observations?

**A:** small values of base give bigger values

**Q:** Why do you think we used $1.1$ in the question rather than $1$ for comparison?

**A:** because for 1 log wil be undefined

#### Activity 3: Plot a graph with $y=\{\log_e x,x,x^2\}$

Calculate corresponding values for $y=\{\log_e x,x,x^2\}$
and store them in variables `y1`, `y2`, and `y3`respectively.

```
# Evaluate log(x) for corresponding values of x.
y1_new = []
for i in range(len(x)):
  y1_new.append(math.log(x[i], np.exp(1)))
y2_new = []
for i in range(len(x)):
  y2_new.append(x[i])
y3_new = []
for i in range(len(x)):
  y3_new.append((x[i]*x[i]))

plt.title("Plots of Logs")
plt.plot(x,y1_new,color = 'red')
plt.plot(x,y2_new,color = 'orange')
plt.plot(x,y3_new,color = 'yellow')
plt.xlabel('Values')
plt.ylabel('log of Values')
plt.show()
```

**Q:** What are your observations?

**A:** X^2 gets very big at large values compared to lnx and x
